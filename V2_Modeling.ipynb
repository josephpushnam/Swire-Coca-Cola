Creating Summary df

```python
# Step 1: Copy original with YEAR for later aggregation
df_with_year = df.copy()

# Step 2: Drop unneeded time columns from main df
df = df.drop(columns=['TRANSACTION_DATE', 'WEEK', 'YEAR'])

# Step 3: Sum Y1_/Y2_ flags
sum_cols = [col for col in df.columns if col.startswith('Y1_') or col.startswith('Y2_')]

# Step 4: All other columns (excluding CUSTOMER_NUMBER) are categorical
cat_cols = [col for col in df.columns if col not in sum_cols + ['CUSTOMER_NUMBER']]

# Step 5: Group main dataframe
grouped = df.groupby('CUSTOMER_NUMBER').agg(
    {**{col: 'sum' for col in sum_cols},
     **{col: 'first' for col in cat_cols}}
).reset_index()

# --------------------------------------------
# Step 6: Add Year-Based Aggregates (ordered + delivery)
# --------------------------------------------
# Columns to sum by year (must match casing in dataset)
extra_sum_cols = [
    'ORDERED_CASES', 'ORDERED_GALLONS',
    'Case_delivery_costs', 'Gallon_delivery_costs'
]

# Step 6.1: Extract necessary columns
df_yearly = df_with_year[['CUSTOMER_NUMBER', 'YEAR'] + extra_sum_cols].copy()

# Step 6.2: Group by customer/year, then pivot year
grouped_year = df_yearly.groupby(['CUSTOMER_NUMBER', 'YEAR']).sum().reset_index()
pivoted_year = grouped_year.pivot(index='CUSTOMER_NUMBER', columns='YEAR')

# Step 6.3: Flatten columns and rename
pivoted_year.columns = [f"Y{1 if year == 2023 else 2}_{col}" for col, year in pivoted_year.columns]
pivoted_year = pivoted_year.reset_index()

# Step 6.4: Create calculated columns
pivoted_year['Y1_Total_Orders'] = (
    pivoted_year.get('Y1_ORDERED_CASES', 0) + pivoted_year.get('Y1_ORDERED_GALLONS', 0)
)
pivoted_year['Y2_Total_Orders'] = (
    pivoted_year.get('Y2_ORDERED_CASES', 0) + pivoted_year.get('Y2_ORDERED_GALLONS', 0)
)
pivoted_year['Y1_Total_Delivery_Costs'] = (
    pivoted_year.get('Y1_Case_Delivery_Costs', 0) + pivoted_year.get('Y1_Gallon_Delivery_Costs', 0)
)
pivoted_year['Y2_Total_Delivery_Costs'] = (
    pivoted_year.get('Y2_Case_Delivery_Costs', 0) + pivoted_year.get('Y2_Gallon_Delivery_Costs', 0)
)

# --------------------------------------------
# Step 7: Merge to get the final result
# --------------------------------------------
final_df = grouped.merge(pivoted_year, on='CUSTOMER_NUMBER', how='left')

# Step 8: View result
final_df.head()
```

### Elbow Method

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Try a range of K values
k_range = list(range(1, 11))
accuracies = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_k = knn.predict(X_test)
    acc = accuracy_score(y_test, y_pred_k)
    accuracies.append(acc)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_range, accuracies, marker='o', linestyle='-', color='blue')
plt.title("Elbow Method for Optimal K in KNN")
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Accuracy on Test Set")
plt.xticks(k_range)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ✅ Function to add R-style significance stars
def add_significance_stars(ols_result):
    summary_df = ols_result.summary2().tables[1].copy()

    def stars(p):
        if p < 0.001:
            return '***'
        elif p < 0.01:
            return '**'
        elif p < 0.05:
            return '*'
        elif p < 0.1:
            return '.'
        else:
            return ''

    summary_df['Significance'] = summary_df['P>|t|'].apply(stars)
    return summary_df

# ✅ Generate summary table
summary_with_stars = add_significance_stars(ols_model)

# ✅ Display top results in Colab
print("📊 OLS Regression Coefficients with R-style Significance:\n")
from IPython.display import display
display(summary_with_stars)
```

```python
# STEP: Model Comparison Setup
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.preprocessing import LabelEncoder

# ✅ Load and filter the dataset (only purchasers)
df = df[df['Y2_Total_Orders'] > 0].copy()

# Drop leakage and irrelevant columns
df_model = df.drop(columns=drop_cols, errors='ignore')

# Encode categoricals
for col in df_model.select_dtypes(include='object').columns:
    df_model[col] = LabelEncoder().fit_transform(df_model[col].astype(str))

# Create features and target
df_model['high_performer'] = (df['Y2_Total_Orders'] >= 615).astype(int)
#X = df_model.drop(columns=['Y2_Total_Orders', 'high_performer'])
y = df_model['high_performer']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

# Define models including MLP
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, class_weight='balanced'),
    "Random Forest": RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42),
    "Support Vector Machine": SVC(probability=True, class_weight='balanced', random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', scale_pos_weight=5, random_state=42),
    "Neural Network (MLP)": MLPClassifier(hidden_layer_sizes=(64, 64), max_iter=500, random_state=42)
}

# Train and evaluate all models
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else y_pred

    results.append({
        "Model": name,
        "Accuracy": round(accuracy_score(y_test, y_pred), 4),
        "Precision": round(precision_score(y_test, y_pred), 4),
        "Recall": round(recall_score(y_test, y_pred), 4),
        "F1 Score": round(f1_score(y_test, y_pred), 4),
        "AUC Score": round(roc_auc_score(y_test, y_prob), 4)
    })

# Output as a DataFrame
results_df = pd.DataFrame(results).sort_values(by="F1 Score", ascending=False)
print("📊 Model Performance Comparison (with MLP Neural Network):")
display(results_df)
```

```python
# ✅ Function to add R-style significance stars
def add_significance_stars(ols_result):
    summary_df = ols_result.summary2().tables[1].copy()

    def stars(p):
        if p < 0.001:
            return '***'
        elif p < 0.01:
            return '**'
        elif p < 0.05:
            return '*'
        elif p < 0.1:
            return '.'
        else:
            return ''

    summary_df['Significance'] = summary_df['P>|t|'].apply(stars)
    return summary_df

# ✅ Generate summary table
summary_with_stars = add_significance_stars(ols_model)

# ✅ Display top results in Colab
print("📊 OLS Regression Coefficients with R-style Significance:\n")
from IPython.display import display
display(summary_with_stars)
```

```python
# Load and clean
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Capstone Projects/Swire Coca Cola/transactional_data.csv')
df['TRANSACTION_DATE'] = pd.to_datetime(df['TRANSACTION_DATE'])

# Weekly aggregation: total gallons ordered across all order types
df['WEEK_START'] = df['TRANSACTION_DATE'].dt.to_period('W').apply(lambda r: r.start_time)
weekly_orders = df.groupby('WEEK_START')['ORDERED_GALLONS'].sum().reset_index()

# Plot it
weekly_orders.set_index('WEEK_START').plot(figsize=(12,4), title="Weekly Ordered Gallons (All Sources)")
```

```python
from causalimpact import CausalImpact

# Create a dummy predictor (flat trend for simplicity)
weekly_orders['predictor'] = range(len(weekly_orders))

# Time series format: [response, predictor]
ts = weekly_orders[['ORDERED_GALLONS', 'predictor']]

# Identify pre/post: Year 1 vs Year 2
# Adjust this if you know the exact launch week of MYCOKE360
pre_period = [0, 51]  # 52 weeks for Year 1
post_period = [52, len(ts)-1]
```

```python
# Set time index for CausalImpact
ts.index = pd.to_datetime(weekly_orders['WEEK_START'])

# Now rerun the model
ci = CausalImpact(ts, pre_period, post_period)
ci.plot()
print(ci.summary())
print(ci.summary(output='report'))
```
